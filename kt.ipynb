```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro constante
c_a = 5

# Crear una malla de valores para x_a y y_a
x_a = np.linspace(0, 10, 100)  # Valores de x_a entre 0 y 10
y_a = np.linspace(0, 10, 100)  # Valores de y_a entre 0 y 10
X_a, Y_a = np.meshgrid(x_a, y_a)  # Crear una malla de coordenadas

# Calcular los valores de la función objetivo
F = (X_a - c_a)**2 + (Y_a - c_a)**2

# Graficar la función objetivo
plt.figure(figsize=(10, 8))
contour = plt.contourf(X_a, Y_a, F, levels=50, cmap='viridis')  # Contornos llenos
plt.colorbar(contour)  # Barra de color para mostrar los valores de la función
plt.title('Gráfica de la Función Objetivo')
plt.xlabel('$x_a$')
plt.ylabel('$y_a$')
plt.scatter(c_a, c_a, color='red', marker='o', label='Punto (c_a, c_a)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros constantes
a_x = 3  # Valor constante para a_x
a_y = 7  # Valor constante para a_y

# Crear una malla de valores para x_a y y_a
x_a = np.linspace(0, 10, 100)  # Valores de x_a entre 0 y 10
y_a = np.linspace(0, 10, 100)  # Valores de y_a entre 0 y 10
X_a, Y_a = np.meshgrid(x_a, y_a)  # Crear una malla de coordenadas

# Calcular los valores de la función objetivo
F = (X_a - a_x)**2 + (Y_a - a_y)**2

# Graficar la función objetivo en 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Superficie
surf = ax.plot_surface(X_a, Y_a, F, cmap='viridis', edgecolor='none')
fig.colorbar(surf)  # Barra de color para mostrar los valores de la función

# Etiquetas y título
ax.set_title('Gráfica de la Función Objetivo en 3D')
ax.set_xlabel('$x_a$')
ax.set_ylabel('$y_a$')
ax.set_zlabel('$f(x_a, y_a)$')

# Marcar el punto (a_x, a_y)
ax.scatter(a_x, a_y, (a_x - a_x)**2 + (a_y - a_y)**2, color='red', s=100, label='Punto (a_x, a_y)')
ax.legend()

plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parámetros constantes
a_x = 3  # Valor constante para a_x
a_y = 7  # Valor constante para a_y

# Crear una malla de valores para x_a y y_a
x_a = np.linspace(0, 10, 100)  # Valores de x_a entre 0 y 10
y_a = np.linspace(0, 10, 100)  # Valores de y_a entre 0 y 10
X_a, Y_a = np.meshgrid(x_a, y_a)  # Crear una malla de coordenadas

# Calcular los valores de la función objetivo
F = (X_a - a_x)**2 + (Y_a - a_y)**2

# Crear gráfico interactivo con Plotly
fig = go.Figure(data=[go.Surface(z=F, x=X_a, y=Y_a, colorscale='Viridis')])

# Añadir detalles al gráfico
fig.update_layout(
    title='Gráfica Interactiva de la Función Objetivo',
    scene=dict(
        xaxis_title='x_a',
        yaxis_title='y_a',
        zaxis_title='f(x_a, y_a)',
    ),
)

# Marcar el punto (a_x, a_y)
fig.add_trace(go.Scatter3d(
    x=[a_x],
    y=[a_y],
    z=[(a_x - a_x)**2 + (a_y - a_y)**2],
    mode='markers',
    marker=dict(size=8, color='red'),
    name='Punto (a_x, a_y)'
))

# Mostrar el gráfico
fig.show()
```

```python
from scipy.optimize import minimize

# Definimos la función objetivo
def objective_function(variables):
    x, y = variables
    return x**2 + y**2  # Ejemplo de función objetivo (minimizar x^2 + y^2)

# Restricciones
constraints = [
    {'type': 'ineq', 'fun': lambda variables: variables[0]},  # x >= 0
    {'type': 'ineq', 'fun': lambda variables: variables[1]},  # y >= 0
    {'type': 'ineq', 'fun': lambda variables: -variables[0] - variables[1]}  # x + y <= 0
]

# Punto inicial
initial_guess = [0, 0]

# Resolviendo el problema
result = minimize(objective_function, initial_guess, constraints=constraints)

# Mostrar resultados
if result.success:
    print("Solución encontrada:")
    print("x =", result.x[0])
    print("y =", result.x[1])
    print("Valor mínimo de la función objetivo:", result.fun)
else:
    print("No se encontró una solución factible.")
```

```python
from sympy import symbols, diff, solve, Eq

# Variables
x, y, lambda1, lambda2, lambda3 = symbols('x y lambda1 lambda2 lambda3', real=True, nonnegative=True)
C = symbols('C', real=True, positive=True)  # Parámetro positivo

# Función objetivo
f = x**2 + y**2  # Ejemplo: minimizar x^2 + y^2

# Lagrangiano
L = f + lambda1 * (-x) + lambda2 * (-y) + lambda3 * (x + y - C)

# Condiciones KKT
stationarity = [diff(L, x), diff(L, y)]
complementary_slackness = [lambda1 * x, lambda2 * y, lambda3 * (x + y - C)]
primal_feasibility = [x >= 0, y >= 0, x + y <= C]

# Resolver las ecuaciones simbólicamente
solutions = solve(stationarity + complementary_slackness + primal_feasibility, [x, y, lambda1, lambda2, lambda3])

# Mostrar solución
if solutions:
    print("Soluciones encontradas:")
    for sol in solutions:
        print(sol)
else:
    print("No se encontraron soluciones.")
```
