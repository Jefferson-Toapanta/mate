```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Definimos la función para graficar
def plot_function(r): #este es el nombre de la función
    x = np.linspace(-3, 3, 400)
    y = r * x + x**3 - x**5
    line.set_ydata(y)
    ax.set_title(f'Gráfica de {r}*x + x^3 - x^5')
    fig.canvas.draw_idle()

# Creamos la figura y el eje
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

# Definimos el rango de x y graficamos la función inicial con r=1
x = np.linspace(-3, 3, 400)
r_initial = 1
y = r_initial * x + x**3 - x**5
line, = ax.plot(x, y, lw=2)

# Configuramos los límites del gráfico
ax.set_xlim([-3, 3])
ax.set_ylim([-10, 10])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

# Creamos el slider para el parámetro r
axcolor = 'lightgoldenrodyellow'
ax_r = plt.axes([0.2, 0.1, 0.65, 0.03], facecolor=axcolor)
slider_r = Slider(ax_r, 'Valor de r', -5.0, 5.0, valinit=r_initial)

# Función de actualización del slider
def update(val):
    r = slider_r.val
    plot_function(r)

slider_r.on_changed(update)

# Mostramos el gráfico
plt.show()
```
