```python
import numpy as np
import plotly.graph_objects as go

# Generar datos
x = np.linspace(0.1, 10, 100)
y = np.linspace(0.1, 10, 100)
x, y = np.meshgrid(x, y)
rho = 1
z = (x**rho + y**rho)**(1/rho)

# Crear la figura
fig = go.Figure(data=[
    go.Surface(
        z=z,
        x=x,
        y=y,
        colorscale='Viridis'
    )
])

# Actualizar el diseo
fig.update_layout(
    title='3D Surface Plot',
    autosize=False,
    width=800,
    height=800,
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Mostrar la figura
fig.show()
```
