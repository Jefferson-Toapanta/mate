```python
import numpy as np

import matplotlib.pyplot as plt

from ipywidgets import interact, FloatSlider



def plot_function(r):

  # Parámetros de la función

  k = 100 # Capacidad de carga (puedes ajustar este valor)

  N = np.linspace(0, 150, 500) # Rango de N (población)

   

  # Función logística

  f = r * N * (1 - (N / k))

   

  # Crear el gráfico

  plt.figure(figsize=(8, 6))

  plt.plot(N, f, label=f"f(N) = {r}N(1 - N/{k})")

  plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)

  plt.axvline(k, color='red', linestyle='--', label=f'Capacidad de carga (k = {k})')

  plt.xlabel("Población (N)")

  plt.ylabel("f(N)")

  plt.title("Modelo Logístico: f(N) = rN(1 - N/k)")

  plt.legend()

  plt.grid()

  plt.show()



# Crear el slider para el parámetro r

interact(plot_function, r=FloatSlider(value=0.5, min=0.1, max=2.0, step=0.1, description='r'));
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Definir la función
def plot_function(r):
    x = np.linspace(-10, 10, 500)  # Valores de x
    y = r + x**2  # Función r + x^2

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'$r + x^2$ con r={r:.2f}')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Línea en y=0
    plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Línea en x=0
    plt.title('Gráfica de $r + x^2$', fontsize=16)
    plt.xlabel('x', fontsize=14)
    plt.ylabel('y', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.show()

# Crear el deslizador y la interacción
slider = FloatSlider(value=0, min=-10, max=10, step=0.1, description='r:')
interact(plot_function, r=slider)
```
