```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin Cobb-Douglas
def u(x1, x2):
    return x1**0.4 * x2**0.6

# Crear un rango de valores para x1 y x2
x1 = np.linspace(0.1, 10, 100)  # Evitar 0 para evitar problemas con la funcin
x2 = np.linspace(0.1, 10, 100)

# Crear una malla de valores
X1, X2 = np.meshgrid(x1, x2)
Z = u(X1, X2)

# Crear la grfica en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='rainbow')

# Etiquetas de los ejes
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('u(x1, x2)')
ax.set_title('Grfica de la funcin Cobb-Douglas')

# Mostrar la grfica
plt.show()
```
