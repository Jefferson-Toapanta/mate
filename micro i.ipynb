```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Rango de valores de x
x_values = np.linspace(-10, 10, 400)

# Función para calcular y verificar la condición
def calculate_y(r):
    return r - x_values**2  # Reorganizamos a y = x² + r

# Función para graficar
def plot_equation(r):
    y_values = calculate_y(r)
    
    plt.figure(figsize=(10, 6))
    
    # Graficar la parábola
    plt.plot(x_values, y_values, label=f'y = x² + {r}', color='blue')
    
    # Graficar el eje X y Y
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    
    # Indicar las raíces si r es negativo
    if r < 0:
        root1 = np.sqrt(-r)
        root2 = -np.sqrt(-r)
        plt.scatter([root1, root2], [0, 0], color='red', label=f'Raíces: {root1:.2f}, {root2:.2f}')
    
    plt.title('Gráfica de r + x² = 0')
    plt.xlabel('x')
    plt.ylabel('y')
    
    # Establecer límites en el eje X e Y entre -10 y 10
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    
    plt.legend()
    plt.grid()
    plt.show()

# Slider para modificar r
r_slider = FloatSlider(value=-1.0, min=-10.0, max=10.0, step=0.1, description='Valor de r:')
interact(plot_equation, r=r_slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Rango de valores de x
x_values = np.linspace(-10, 10, 400)

# Función para calcular y verificar la condición
def calculate_y(r):
    return -r + x_values**2  # Reorganizamos a y = x² + r

# Función para graficar
def plot_equation(r):
    y_values = calculate_y(r)
    
    plt.figure(figsize=(10, 6))
    
    # Graficar la parábola
    plt.plot(x_values, y_values, label=f'y = x² + {r}', color='blue')
    
    # Graficar el eje X y Y
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    
    # Añadir flechas para indicar estabilidad
    if r < 0:
        # Estable: flecha a la derecha
        plt.annotate('', xy=(8, 20), xytext=(5, 15),
                     arrowprops=dict(facecolor='green', shrink=0.05))
        plt.text(6.5, 18, 'Estable', fontsize=12, color='green')
    elif r > 0:
        # Inestable: flecha a la izquierda
        plt.annotate('', xy=(-8, -20), xytext=(-5, -15),
                     arrowprops=dict(facecolor='red', shrink=0.05))
        plt.text(-6.5, -18, 'Inestable', fontsize=12, color='red')

    plt.title('Gráfica de r + x² = 0')
    plt.xlabel('x')
    plt.ylabel('y')
    
    # Establecer límites en el eje X e Y entre -10 y 10
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    
    plt.legend()
    plt.grid()
    plt.show()

# Slider para modificar r
r_slider = FloatSlider(value=-1.0, min=-10.0, max=10.0, step=0.1, description='Valor de r:')
interact(plot_equation, r=r_slider)
```
